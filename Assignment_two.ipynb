```python
#Question10: Create a program that asks the user to enter a password repeatedly using a loop until the correct password is entered.
correct_password = "python123"

while True:
    password = input("Enter password: ")

    if password == correct_password:
        print("Login Successful")
        break
    else:
        print("Wrong Password, Try Again")
```

```python
# Question9: Write a function that takes two numbers as arguments and returns the larger number using conditional statements.
def larger_number(num1, num2):
    if num1 > num2:
        return num1
    else:
        return num2

result = larger_number(10, 20)
print("Larger Number =", result)
```

```python
#Question5: write a program using logical operators that checks: * If a student’s marks are greater than 50 **and** attendance is greater than 75, print “Allowed in exam” * Otherwise print “Not allowed”
marks = int(input("Enter marks: "))
attendance = int(input("Enter attendance percentage: "))

if marks > 50 and attendance > 75:
    print("Allowed in exam")
else:
    print("Not allowed")
```

```python
# Question 3: Create a program that checks whether a number is positive, negative, or zero using conditional statements.
num = int(input("Enter a number: "))

if num > 0:
    print("Positive Number")
elif num < 0:
    print("Negative Number")
else:
    print("Zero")
```

```python
# Question 2: Write a program that takes two numbers from the user and prints: * Addition * Subtraction * Multiplication * Division
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Addition =", num1 + num2)
print("Subtraction =", num1 - num2)
print("Multiplication =", num1 * num2)

if num2 != 0:
    print("Division =", num1 / num2)
else:
    print("Division by zero is not allowed")
```
